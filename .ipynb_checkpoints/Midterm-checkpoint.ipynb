**Question 3:**

"99 Bottles of Beer" is a traditional song in the United States and Canada. It is popular to sing on long trips, as it has a very repetitive format which is easy to memorize, and can take a long time to sing. The song's simple lyrics are as follows:
99 bottles of beer on the wall, 99 bottles of beer.  
Take one down, pass it around, 98 bottles of beer on the wall.
The same verse is repeated, each time with one fewer bottle. The song is completed when the singer or singers reach zero.
Your task here is to write a program capable of generating all the verses of the song.

```python
i = 99
bottles = []
while i >= 0:
    bottles.append(i)
    i -= 1
for bottle in bottles:
    print bottle,"bottles of beer on the wall,"
    print bottle,"bottles of beer."
    print "Take one down, pass it around," 
    print bottle-1,"bottles of beer on the wall."
    print "\n"
#hilarious
```

**Question 4:**
    
Plot a simple graph of a sine function in the range 0 to 3 with a step size of 0.1.
Make the line red.
Add green diamond-shaped markers with size of 7.
Include legible axis labels.
Add a legend and a grid to the plot.

```python
%pylab inline
#x = np.arange(0, 3, 0.1) #cuts off the graph at x = 3 so I used linspace.
x = np.linspace(0, 3, 30)
y = np.sin(x)
fig, ax1 = plt.subplots(1, 1)
ax1.set_xlabel('x',fontsize=15)
ax1.set_ylabel('y = sin(x)',fontsize=15)
plt.plot(x,y, color='r', label = "sin(x)");
plt.plot(x,y,'D',markersize=7, color='g', label = "sin(x)");
plt.grid()
plt.legend(loc='upper left')
```
