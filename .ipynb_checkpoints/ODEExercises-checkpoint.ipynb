```python
g = 9.8 #m/s^2
ell = 0.5 #m
th0 = 20.*np.pi/180 # 20 degrees, but must use radians
th1 = 0
th2 = np.pi/4
th3 = np.pi/2
th4 = np.pi*3/4
th5 = np.pi

t = np.arange(0.,10.+0.01,0.01)
theta_app1 = lambda t: th1*np.cos(np.sqrt(g/ell)*t)
theta_app2 = lambda t: th2*np.cos(np.sqrt(g/ell)*t)
theta_app3 = lambda t: th3*np.cos(np.sqrt(g/ell)*t)
theta_app4 = lambda t: th4*np.cos(np.sqrt(g/ell)*t)
theta_app5 = lambda t: th5*np.cos(np.sqrt(g/ell)*t)

plt.plot(t,theta_app1(t),label=r'$\theta (rad)=0$')
plt.plot(t,theta_app2(t),label=r'$\theta (rad)= \pi/4$')
plt.plot(t,theta_app3(t),label=r'$\theta (rad)= \pi/2$')
plt.plot(t,theta_app4(t),label=r'$\theta (rad)= 3\pi/4$')
plt.plot(t,theta_app5(t),label=r'$\theta (rad)= \pi$')

plt.xlabel(r'$t (seconds)$',fontsize=20)
plt.ylabel(r'$\theta (rad)$',fontsize=20)
max_app = max(max(theta_app1(t)),max(theta_app2(t)),max(theta_app3(t)),max(theta_app4(t)),max(theta_app1(t)))
plt.ylim(-max_app-1,max_app+1)
plt.grid()
legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
def deriv(theta_current,t):  # return derivatives of the array theta
    g = 9.8 #m/s^2
    ell = 0.5 #m
    theta = theta_current[0]
    omega = theta_current[1]
    dtheta = omega
    domega = -(g/ell)*np.sin(theta)
    return np.array([ dtheta, domega ])
```

```python
t = np.arange(0.,10.+0.01,0.01)
th1 = 0
th2 = np.pi/4
th3 = np.pi/2
th4 = np.pi*3/4
th5 = np.pi

omega0 = 0. 
theta_init1 = np.array([th1, omega0]) 
theta_init2 = np.array([th2, omega0])
theta_init3 = np.array([th3, omega0])
theta_init4 = np.array([th4, omega0])
theta_init5 = np.array([th5, omega0])

theta1 = odeint(deriv,theta_init1,t)
theta2 = odeint(deriv,theta_init2,t) 
theta3 = odeint(deriv,theta_init3,t) 
theta4 = odeint(deriv,theta_init4,t) 
theta5 = odeint(deriv,theta_init5,t)

plt.plot(t,theta1[:,0],'r-',label= r'$\theta (rad)=0$')
plt.plot(t,theta2[:,0],'b-',label=r'$\theta (rad)= \pi/4$')
plt.plot(t,theta3[:,0],'g-',label=r'$\theta (rad)= \pi/2$')
plt.plot(t,theta4[:,0],'y-',label=r'$\theta (rad)= 3\pi/4$')
plt.plot(t,theta5[:,0],'c-',label=r'$\theta (rad)= \pi$')

plt.xlabel(r'$t (seconds)$',fontsize=20)
plt.ylabel(r'$\theta (rad)$',fontsize=20)

max_app = max(max(theta1[:,0]),max(theta2[:,0]),max(theta3[:,0]),max(theta4[:,0]),max(theta5[:,0]))

plt.ylim(-max_app-1,max_app+1)
plt.grid()
legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

The period of oscillation is stretched out and out of phase with the other waves, whereas in the approximation above they period is consistent. When theta = 0 the pendulum is not displaced initially so it will not oscillate. When theta = pi it just spins around and around and around.

# Question 2

(a) Use Euler's method (not odeint) to solve the following first-order differential equation with given initial condition y(0) = 0.
\begin{equation}
\frac{dy}{dx} = x + 2y
\end{equation}

Find the value of y when x = 1, y(1), using two different step sizes: 0.25 then 0.02.

```python
def EulersMethod(step,val):
    xmax = 10 # maximum x
    dx = step # x step
    nsteps = int(xmax/dx)
    x = np.zeros(nsteps)
    y = np.zeros(nsteps)
    dy = np.zeros(nsteps)
    # set initial conditions, redundant in this example

    for i in range(len(x)-1):
        x[i+1] = x[i] + dx
        dy[i+1] = x[i] + 2*y[i]
        y[i+1] = y[i] + dx * dy[i]
    #index = np.where(x==val) not sure why this doesn't work when dx = 0.02 ...
    index = val/step
    return x,y,y[index]
```

```python
x, y, y1 = EulersMethod(0.25,1)
print "y(1) =", y1
```

```python
x, y, y1 = EulersMethod(0.02,1.)
print "y(1) =", y1
```

(b) Plot the results from part (a) for the two different step sizes with the exact solution
\begin{equation}
y = 0.25 e^{2x} - 0.5 x - 0.25
\end{equation}
on one plot. Include a legend. Which step size gives a more accurate result?

```python
x, y, y1 = EulersMethod(0.25,1)
plt.plot(x,y, label="Step size = 0.25")
x, y, y1 = EulersMethod(0.02,1)
plt.plot(x,y, label="Step size = 0.02")
y = lambda x: 0.25*exp(2*x) - 0.5*x - 0.25
plt.plot(x,y(x),label = "Exact Solution")
plt.xlabel(r'$x$',size = 20)
plt.ylabel(r'$y$',size = 20)
plt.ylim(0,4)
plt.xlim(0,4)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

The approximation with finer resolution (step = 0.02) is more accurate.

```python
def EulerCromerMethod(step,val=1):
    xmax = 10 # maximum time
    dx = step # time step
    nsteps = int(xmax/dx)
    dy = np.zeros(nsteps)
    x = np.zeros(nsteps)
    y = np.zeros(nsteps)

    for i in range(len(x)-1):
        x[i+1] = x[i] + dx
        dy[i+1] = x[i] + 2*y[i]
        y[i+1] = y[i] + dx * dy[i+1]
        
    return x,y
```

```python
def MidpointMethod(step,val=1):
    xmax = 10 # maximum time
    dx = step # time step
    nsteps = int(xmax/dx)
    dy = np.zeros(nsteps)
    x = np.zeros(nsteps)
    y = np.zeros(nsteps)
    for i in range(len(x)-1):
        x[i+1] = x[i] + dx
        dy[i+1] = x[i] + 2*y[i]
        y[i+1] = y[i] + dx * 0.5*(dy[i]+dy[i+1])
        
    return x,y
```

```python
x,y = EulerCromerMethod(0.25)
```

```python
x,y = MidpointMethod(0.25)
```

```python
x1, y1 = EulerCromerMethod(0.25)
plt.plot(x1,y1, label="ECM Step size = 0.25")
x2, y2 = EulerCromerMethod(0.02)
plt.plot(x2,y2, label="ECM Step size = 0.02")
y = lambda x: 0.25*exp(2*x) - 0.5*x - 0.25
plt.plot(x2,y(x2),label = "Exact Solution")
diff1 = abs(y1-y(x1))
diff2 = abs(y2-y(x2))
plt.plot(x1,diff1,label = "ECM Step = 0.25 difference")
plt.plot(x2,diff2,label = "ECM Step = 0.02 difference")

plt.xlabel(r'$x$',size = 20)
plt.ylabel(r'$y$',size = 20)
plt.ylim(0,4)
plt.xlim(0,4)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
x1, y1 = MidpointMethod(0.25)
plt.plot(x1,y1, label="Midpoint Method Step size = 0.25")
x2, y2 = MidpointMethod(0.02)
plt.plot(x2,y2, label="Midpoint Method Step size = 0.02")
y = lambda x: 0.25*exp(2*x) - 0.5*x - 0.25
plt.plot(x2,y(x2),label = "Exact Solution")
diff1 = abs(y1-y(x1))
diff2 = abs(y2-y(x2))
plt.plot(x1,diff1,label = "Midpoint Method Step = 0.25 difference")
plt.plot(x2,diff2,label = "Midpoint Method Step = 0.02 difference")

plt.xlabel(r'$x$',size = 20)
plt.ylabel(r'$y$',size = 20)
plt.ylim(0,4)
plt.xlim(0,4)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

(a) Create a derivative function to solve this system of equations and use odeint to solve it for a projectile fired at an angle of 50 degrees from the horizontal at an initial speed of 30 m/s. Track the projectile's motion over the time interval of 0 to 200 seconds and let (x0,y0) = (0,0). You'll have to calculate (vx0,vy0) from the information provided. Let vter=44 m/s. Plot the trajectory of the projectile (y vs. x) to see how far it can travel in the x direction before landing.

```python
def DragDeriv(inits,t):
    termv = inits[4]
    vx0 = inits[0]
    vy0 = inits[1]
    x = inits[2]
    y = inits[3]
    g = 9.8
    
    dvx = -g*vx0/termv**2*sqrt(vx0**2+vy0**2)
    dvy = -g*(1+vy0/termv**2*sqrt(vx0**2+vy0**2))
    
    return np.array([dvx,dvy,vx0,vy0])
```

```python
def interactivePlot(theta,V,T,x0,y0,termv):
    v0 = [V*np.cos(theta*np.pi/180),V*np.sin(theta*np.pi/180)]

    t = np.arange(0.,T+0.1,0.1)

    soln = odeint(DragDeriv,np.array([v0[0],v0[1],x0,y0,termv]),t)

    plt.plot(soln[:,2],soln[:,3])
    plt.xlim(0,80)
    plt.ylim(0,max(soln[:,3])+10)
    plt.xlabel(r'$X$',fontsize = 20)
    plt.ylabel(r'$Y$',fontsize = 20)
    plt.show()

widg = interact(interactivePlot,theta=(0.,np.pi),V = (0.,60.),x0 = (0.,10.),y0 = (0.,10.),termv = (0.,60))
```
