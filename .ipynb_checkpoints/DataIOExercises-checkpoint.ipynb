```python
fout = open('clean_data.dat', 'w')
for line in open('messy_data.dat'):
    for item in line.split(','): #str.split(str="", num=string.count(str)).
        fout.write('%8.5f'%float(item))
        fout.write('\t') #\t denotes a "tab" on the keyboard
    fout.write('\n')
fout.close()
#split() works like this:
#str = "Line1-abcdef \nLine2-abc \nLine4-abcd";
#print str.split( );
#print str.split(' ', 1 );
#['Line1-abcdef', 'Line2-abc', 'Line4-abcd']
#['Line1-abcdef', '\nLine2-abc \nLine4-abcd']
```

```python
data = np.loadtxt("messy_data.dat", delimiter=',')
np.savetxt("clean_data.dat", data, delimiter='\t', fmt="%8.5f")
#numpy.savetxt(fname, X, fmt='%.18e', delimiter=' ', newline='\n', header='', footer='', comments='# ')[source]¶
```
